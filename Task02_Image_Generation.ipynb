```python
from PIL import Image, ImageDraw

# Create blank image
img = Image.new("RGB", (512, 512), color="white")

# Draw text
draw = ImageDraw.Draw(img)

draw.text(
    (50, 200),
    "Artificial Intelligence\nGenerated Image",
    fill="black"
)

# Save image
img.save("generated_image.png")

# Display image
img
```
